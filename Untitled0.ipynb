```python
def get_ds_split(dataset,train=0.8,test=0.1,val=0.1,shuffle_size=10000,shuffle=True):
  if shuffle:
    dataset=dataset.shuffle(shuffle_size,seed=12)

  train_size = int(len(dataset)*0.8)
  train_data = dataset.take(train_size)
  remain= dataset.skip(train_size)
  val_size = int(len(dataset)*0.1)
  val_data = remain.take(val_size)
  test_data = remain.skip(val_size)

  return train_data, val_data, test_data
```

```python
train_dataseet , validate_dataset , test_dataset=get_ds_split(potato_ds)
```

```python
train_ds=train_dataseet.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds=test_dataset.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds=validate_dataset.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
from keras.layers import Resizing, Rescaling
from keras.layers import RandomFlip, RandomRotation, RandomZoom
import keras


resize_and_scale = keras.Sequential([
    Resizing(256, 256),
    Rescaling(1.0/255)
])

augmentation = keras.Sequential([
    RandomFlip("horizontal"),
    RandomRotation(0.2)
])
```

```python
from keras import layers
```

```python
from keras import Input

model=keras.Sequential([
    Input(shape=(256,256,3)),
    resize_and_scale,
    augmentation,
    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPool2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPool2D((2,2)),
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPool2D((2,2)),
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPool2D((2,2)),
    layers.Flatten(),
    layers.Dense(3, activation='softmax'),
])

```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='SparseCategoricalCrossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(train_ds,epochs=25, batch_size=32,verbose=1, validation_data= val_ds)
```

```python
score=model.evaluate(test_ds)
```
